## K Means Cluster Creation

Now it is time to create the Cluster labels!

** Import KMeans from SciKit Learn.**

```python
from sklearn.cluster import KMeans
```

```python
km =  KMeans(n_clusters=2)
X=college.drop('Private', axis=1)
Y=college['Private']
km.fit(X)
```

```python
km.cluster_centers_
#evaluating cluster centers
```

```python
km.labels_
```

## Evaluation

There is no perfect way to evaluate clustering if you don't have the labels, however since this is just an exercise, we do have the labels, so we take advantage of this to evaluate our clusters, keep in mind, you usually won't have this luxury in the real world.

** Create a new column for df called 'Cluster', which is a 1 for a Private school, and a 0 for a public school.**

```python
#college.drop('Cluster',axis=1, inplace=True)
college['KMCluster_Pred'] = km.labels_#predicted labels
def convert_class(str):
    if(str=='Yes'):
        return 1
    else:
        return 0
college['Cluster_Actual']=college['Private'].apply(lambda x: convert_class(x))#actual labels
```

```python
college.head()
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print('Classification Report')
print(classification_report(college['KMCluster_Pred'],college['Cluster_Actual']))
print('Confusion Matrix')
print(confusion_matrix(college['KMCluster_Pred'], college['Cluster_Actual']))
```

```python
#K-Means clustering, given the nature of unsupervised algorithms containing no labels to learn and a priori info, has performed well.
#Especially, considering that it only clusters around random centers without a priori info., it classifies private and public universities well.
```
